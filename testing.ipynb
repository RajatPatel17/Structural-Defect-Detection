```python
import google.generativeai as genai
```

```python
from PIL import Image
import os
```

```python
defect_image = Image.open('defect.jpg')
```

```python
gemini_api_key = os.getenv('Google_API_KEY2')
genai.configure(api_key=gemini_api_key)
model = genai.GenerativeModel('gemini-2.5-flash-lite')
```

```python
prompt = 'Identify the structural defects in the image and give tell which defect it is and possible remedies'
```

```python
response = model.generate_content([prompt,defect_image],generation_config={'temperature':0.7})
```

```python
print(response.text)
```
